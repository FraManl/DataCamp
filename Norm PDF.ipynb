```python
import pandas as pd
import numpy as np
import os, sys
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

a = [0, 1]
k = 6
n = .01

dataset = pd.DataFrame(
    {
        'T1': np.arange(-k, k, n),
        'T2': np.arange(-k, k, n),
    }
)

dataset['pdf'] = norm.pdf(dataset['T1'])
dataset['cdf'] = norm.cdf(dataset['T1'])

sns.set()
sns.set_style('white')

sns.lineplot(x="T1", y="pdf", data=dataset)
plt.show()
sns.lineplot(x="T1", y="cdf", data=dataset)
plt.show()
```
