```python
import pandas as pd
import numpy as np
import os, sys
import glob
import time
from pathlib import Path

prefix = str(Path.home())
suffix = "\Desktop\_Dashboard JIRA\data"
path = prefix + suffix
os.chdir(path)

extension = "/*.csv"
CSV_files = glob.glob(os.getcwd() + extension)
latest_CSV_file = max(CSV_files, key = os.path.getctime)
jira_df = pd.read_csv(latest_CSV_file, sep='|', encoding='utf-8')

StoryPoints = 'Champs personnalisés (Story Points)'
EpicLink = 'Champs personnalisés (Epic Link)'
EpicName= 'Champs personnalisés (Epic Name)'

jira_df = jira_df.rename(columns={StoryPoints : 'StoryPoints'})
jira_df = jira_df.rename(columns={EpicLink: 'EpicLink'})
jira_df = jira_df.rename(columns={EpicName: 'EpicName'})

jira_df = jira_df[jira_df.columns.drop(list(jira_df.filter(regex='Champs personnalisés')))]

jira_df
```
